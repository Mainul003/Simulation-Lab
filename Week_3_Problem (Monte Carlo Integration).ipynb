```python
# Package management
import math
from scipy import random
import numpy as np
import matplotlib.pyplot as plt
```

## Compute using monte carlo integration
$$\int_0^\pi sin(x)  \,dx $$

```python
#Define Function
def func(x):
    return np.sin(x)
    
```

```python
#Define The limit
a = 0
b = np.pi


#Define Sample for the sum
N = 1000
xrand = random.uniform(a,b,N)

# Calculate the Integral
integral = 0.0
for i in range(N):
    integral += func(xrand[i])
result = (b-a)*(1/N)*integral

print(result)
    

```
